## Volatility 

```python
returns = stock.returns.Adj_close.values
```

```python
def getVolatility(returns):
    return np.std(returns)
```

```python

m_prev = 0
m_delta = 21
volatility_vect = []
for i in range(len(returns)-m_delta):
    volatility_vect.append(getVolatility(returns[m_prev:m_prev+m_delta]))
    m_prev = m_prev+1
    
```

```python
annual_vol = np.array(volatility_vect)*np.sqrt(360)
```

```python
plt.plot(annual_vol)
```

```python
def getMinMax(vector,sigma_level=3):
    _mean = np.mean(vector)
    _std  = np.std(vector)
    return (_mean-sigma_level*_std if _mean-sigma_level*_std>0 else 0), _mean+sigma_level*_std
```

```python
volatility_interval = getMinMax(annual_vol)
```

```python
volatility_interval
```

## Delimitación del problema

```python
# variables en consideración

variables = {
    
    'strike':{'min':stock.prices.Adj_close.min(),
             'max':stock.prices.Adj_close.max(),
             '_delta':0.01,
             '_type':'numeric'},
    
    'risk_free':{'min':r-0.01,
                 'max':r+0.01,
                 '_delta':0.001,
                 '_type':'numeric'},
    
    'volatility':{'min':volatility_interval[0],
                  'max':volatility_interval[-1],
                  '_delta':0.001,
                  '_type':'numeric'},
    
    'initial_value':{'min':stock.prices.Adj_close.min(),
                     'max':stock.prices.Adj_close.max(),
                     '_delta':0.01,
                     '_type':'numeric'},
    
    'maturity':{'min':1,
                'max':1,
                '_delta':1,
                '_type':'numeric'},
    
    'barrier':{'min':stock.prices.Adj_close.min(),
               'max':stock.prices.Adj_close.max(),
               '_delta':0.01,
               '_type':'numeric'},
    
    'option':{'values':['european_call','european_put'],'_type':'categoric'}
}
```

```python
def randomChoice(nelements):
    rand = np.random.uniform(0,1)
    limits = 1/nelements 
    return int(rand/limits)

def recursiveChoice(n,nelements):
    if n==1:
        return [randomChoice(nelements)]
    n = n-1
    return [randomChoice(nelements)]+recursiveChoice(n,nelements)

def getValues(n,param):
    if 'numeric' in param['_type']:
        return np.random.uniform(low=param['min'],high=param['max'],size=n)
    nelements = len(param['values'])
    return list(map(lambda x: param['values'][x],recursiveChoice(n,nelements)))
    
def createTable(n=10):
    column_name = list(variables.keys())
    df = pd.DataFrame([getValues(n,variables[c]) for c in column_name]).T
    df.columns = column_name
    return df
```

```python
def analyticBlackScholes(_dict):#_type,St,K,r,sigma,T,t):
    _type = _dict['option']
    St = _dict['initial_value']
    K = _dict['strike']
    r = _dict['risk_free']
    sigma = _dict['volatility']
    T = _dict['maturity']
    B = _dict['barrier']
    
    d1 = (np.log(St/K)+(r+sigma**2/2)*(T))/(sigma*np.sqrt(T))
    d2 = d1-sigma*np.sqrt(T)
    d3 = (np.log(St/B)+(r+sigma**2/2)*(T))/(sigma*np.sqrt(T))
    d4 = d3-sigma*np.sqrt(T)
    d5 = (np.log(St/B)-(r-sigma**2/2)*(T))/(sigma*np.sqrt(T))
    d6 = d5-sigma*np.sqrt(T)
    d7 = (np.log(St*K/B**2)-(r-sigma**2/2)*(T))/(sigma*np.sqrt(T))
    d8 = d7-sigma*np.sqrt(T)
    
    # european options 
    if _type=='european_call':
        return St*norm.cdf(d1)-K*np.exp(-r*(T))*norm.cdf(d2)
    if _type=='european_put':
        return -St*norm.cdf(-d1)+K*np.exp(-r*(T))*norm.cdf(-d2)
    
    print('Error: Type not found.')
    return None 
```

```python
def one_hot(attrs, attr):
    def hot(x):
        if x == str(attr):
            return 1
        else:
            return 0
    return [hot(a) for a in attrs]
```

```python
dataset = createTable(100)
dataset['option_value'] = dataset.apply(analyticBlackScholes,1).values
dataset.head()
```

```python
dataset.query('option == "european_call"').plot.scatter(x='volatility',y='option_value',label='european_call',figsize=(10,5))
dataset.query('option == "european_put"').plot.scatter(x='volatility',y='option_value',label='european_put',figsize=(10,5))
plt.show()
```
